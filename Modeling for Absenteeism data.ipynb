```python
import pandas as pd
import numpy as np
```

```python
data_preprocessed = pd.read_csv('Absenteeism_Preprocessed.csv')
data_preprocessed.head()
```

## Preprocessing the Absenteeism Time in Hours

### We will take the median of the column and divide it into two classes, one is moderate absence & Excessive Absence as 0 & 1

```python
Median = data_preprocessed['Absenteeism Time in Hours'].median()
Median
```

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] > Median, 1, 0)
```

```python
data_preprocessed['Targets'] = targets
```

```python
data_preprocessed.head(10)
```

```python
data_preprocessed = data_preprocessed.drop(['Absenteeism Time in Hours'], axis=1)
```

```python
data_targets = data_preprocessed.copy()
```

```python
data_targets.head()
```

```python
#inputs
unscaled_inputs = data_targets.iloc[:,:-1]
targetss = data_targets['Targets']
unscaled_inputs
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = scaler.transform(unscaled_inputs)
```

```python
scaled_inputs
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train, x_test, y_train, y_test = tts(scaled_inputs, targetss, random_state=360, train_size=0.8)
```

# Modeling

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *
```

```python
model = LogisticRegression(verbose=2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
prediction = model.predict(x_train)
```

```python
prediction == y_train
```

```python
np.sum(prediction == y_train)
```

```python
# Therefore accuracy can be obtained by divided by total number of True by no of obs
np.sum(prediction == y_train) / prediction.shape
```

```python
model.coef_
```

```python
model.intercept_
```

```python
summary = pd.DataFrame(data=unscaled_inputs.columns.values, columns=['Feature Names'])
summary['Weights'] = np.transpose(model.coef_)
summary
```

```python
summary.index = summary.index + 1
summary.loc[0] = ['Bias', model.intercept_[0]]
summary = summary.sort_index()
summary
```

```python
summary['Odds_Ratio'] = np.exp(summary.Weights)
summary
```

```python
summary.sort_values('Odds_Ratio', ascending=False)
```

### With the help of odds ratio & weights we can clearly see that Daily work load avg, Day of the week, Education, Distance to work are useless

# Testing the Model

```python
model.score(x_test,y_test)
```
