```python
reasons_dummies = pd.get_dummies(df['Reason for Absence'], drop_first = True)
```

```python
df = df.drop(['Reason for Absence'], axis=1)
```

```python
# max function gives us the maximum value in the axis 1 (row) by this we can find the rows only with zeros
# And we have grouped the dummies into four classes of reasons
reason_type_1 = reasons_dummies.loc[:,1:14].max(axis=1)
reason_type_2 = reasons_dummies.loc[:,15:17].max(axis=1)
reason_type_3 = reasons_dummies.loc[:,18:21].max(axis=1)
reason_type_4 = reasons_dummies.loc[:,22:].max(axis=1)
```

```python
df = pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4], axis=1)
df.head()
```
